disease count using pie chart

```python
disease_count = df['Disease'].value_counts()

plt.figure(figsize=(7, 7))
plt.pie(
    disease_count.values,
    labels=disease_count.index,
    autopct='%1.1f%%'
)

plt.title('Disease Distribution')
plt.show()
```

Age wise patient count with pie chart

```python

bins = [0, 18, 35, 60, 100]
labels = ['Child', 'Young Adult', 'Adult', 'Senior']

df['Age_Group'] = pd.cut(
    df['Age'],
    bins=bins,
    labels=labels
)
age_group_count = df['Age_Group'].value_counts()


print(age_group_count)

plt.figure(figsize=(6, 6))
age_group_count.plot(
    kind='pie',
    autopct='%1.1f%%'
)
plt.title("Patient Distribution by Age Group")
plt.ylabel('')
plt.show()
```

 Patient Admission date calculates by month with bar chart

```python
df['Admission_date']=pd.to_datetime(
    df['Admission_Date']
)
df['month']= \
    df['Admission_date'].dt.month_name()
monthly_admissions=df['month'].value_counts()
print(monthly_admissions)
plt.figure(figsize=(10,5))
monthly_admissions.plot( kind='bar')
plt.title("monthly patient admissions")
plt.xlabel("month")
plt.ylabel("number of patients")
plt.show()
```

```python



df['Admission_Date'] = pd.to_datetime(df['Admission_Date'])
df['Discharge_Date'] = pd.to_datetime(df['Discharge_Date'])


df['Length_of_Stay'] = (
    df['Discharge_Date'] - df['Admission_Date']
).dt.days

print("Length of Stay Statistics")
print("Average Stay:", round(df['Length_of_Stay'].mean(), 2), "days")
print("Maximum Stay:", df['Length_of_Stay'].max(), "days")
print("Minimum Stay:", df['Length_of_Stay'].min(), "days")


plt.figure(figsize=(8,5))
plt.hist(df['Length_of_Stay'], bins=10)

plt.title('Length of Stay Distribution')
plt.xlabel('Length of Stay (Days)')
plt.ylabel('Number of Patients')

plt.show()
```
