```python
import numpy as np
products=["laptop","mobile","tablet","headphone","smartwatch"]
sales=[12000,15000,18000,14000,20000] # Defining 'sales' for this cell as it's used below

print("total sales:",np.sum(sales))
print("average sales:",np.mean(sales))
print("lowest sales:",np.min(sales))
best_index=np.argmax(sales)
print("best selling product:", products[best_index])
print("\n product sales report")
for i in range(len(products)):
    print(products[i],":",sales[i])
```

```python
import pandas as pd
employee={
"emp_id":[101,102,103,104,105],
"name":["arun","priya","rahul","divya","kumar"],
"department":["it","hr","it","finance","hr"],
"salary":[46000,40000,50000,55000,42000],
"experience":[2,3,4,5,6]
}
df=pd.DataFrame(employee)
print("employee dataset")
print(df)

print("average salary:")
print(df["salary"].mean())
print("highest salary:")
print(df["salary"].max())
print("employee in hr:")
print(df[df["department"]=="hr"])
print("employees with experience>3:")
print(df[df["experience"]>3])
print("employee count by department:")
print(df["department"].value_counts())
```
