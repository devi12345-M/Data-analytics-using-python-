```python
Mark1 = int(input("Enter Mark1: "))
Mark2 = int(input("Enter Mark2: "))
Mark3 = int(input("Enter Mark3: "))

Total = Mark1 + Mark2 + Mark3
Average = Total / 3

print("Total :", Total)
print("Average :", Average)

if Average >= 50:
    print("Pass")
else:
    print("Fail")
```
