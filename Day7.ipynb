```python
import pandas as pd
data={
"emp_nm":["john","marry"],
"emp_sal":[25000,40000]
}
df=pd.DataFrame(data)
df.rename(columns={
  "emp_nm":"employee name",
  "emp_sal":"employee salary"
},inplace=True)
print(df)
```

```python
import pandas as pd
data={
"region":["south","north","east","west","south","north"],
"product":["laptop","mobile","laptop","tablet","mobile","tablet"],
"sales":[50000,30000,45000,25000,20000,35000]
}
df=pd.DataFrame(data)
print(df)
```

```python
region_sales=df.groupby("region")["sales"].sum()
print(region_sales)
```

```python
south_sales=df[df["region"]=="south"]
print(south_sales)
```

```python
high_sales=df[df["sales"]>30000]
print(high_sales)
```

```python
import pandas as pd

data={
"emp_nm":["john","marry"],
"emp_sal":[25000,40000]
}
df=pd.DataFrame(data)
df.rename(columns={
  "emp_nm":"employee name",
  "emp_sal":"employee salary"
},inplace=True)

sorted_df=df.sort_values("employee salary")
print(sorted_df)
```

```python
import pandas as pd
data={
"salary":[5000,7000,15000,None]
}
df=pd.DataFrame(data)
mean_value=df["salary"].mean()
df["salary"].fillna(mean_value,inplace=True)
print(df)
```
