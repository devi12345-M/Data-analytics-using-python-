```python

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

data = {
    "OrderID": [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    "Region": [
        "North",
        "South",
        "East",
        "West",
        "North",
        "East",
        "South",
        "West",
        "North",
        "East",
    ],
    "Product": [
        "Laptop",
        "Mobile",
        "Tablet",
        "Laptop",
        "Mobile",
        "Laptop",
        "Tablet",
        "Mobile",
        "Tablet",
        "Laptop",
    ],
    "SalesAmount": [1200, 800, 450, 1300, 850, 1250, 400, 900, 500, 1100],
}

df = pd.DataFrame(data)




region_report = (
    df.groupby("Region")["SalesAmount"]
    .agg(Total_Sales="sum", Average_Sales="mean", Total_Orders="count")
    .reset_index()
)


region_report = region_report.sort_values(by="Total_Sales", ascending=False)


pivot_report = df.pivot_table(
    values="SalesAmount", index="Region", columns="Product", aggfunc="sum"
).fillna(0)


print("=== REGION-WISE SALES SUMMARY ===")
print(region_report.to_string(index=False))
print("\n=== REGION VS PRODUCT SALES (PIVOT TABLE) ===")
print(pivot_report)


region_report.to_csv("region_sales_summary.csv", index=False)
pivot_report.to_csv("region_product_matrix.csv")
print("\nReports successfully exported to CSV.")

plt.figure(figsize=(8, 5))
sns.barplot(
    x="Region",
    y="Total_Sales",
    data=region_report,
    hue="Region",
    palette="viridis",
    legend=False,
)
plt.title("Total Revenue Contribution by Region")
plt.xlabel("Geographic Region")
plt.ylabel("Total Sales ($)")
plt.tight_layout()

plt.savefig("region_sales_chart.png")
plt.show()
```

```python
import matplotlib.pyplot as plt
months=["jan","feb","mar","apr","may","jun"]
sales=[10000,12000,15000,13000,18000,20000]


plt.figure(figsize=(18, 5))

plt.subplot(1, 3, 1)
plt.plot(months, sales, marker="o", linestyle="-", color="green")
plt.title("Sales Trend")
plt.xlabel("Months")
plt.ylabel("Sales")


plt.subplot(1, 3, 2)
plt.bar(months,
        sales,
        color="orange")
plt.title("Sales Comparison")
plt.xlabel("Months")
plt.ylabel("Sales")


expenses=[5000,3000,2000]
categories=["food","travel","shopping"]


plt.subplot(1, 3, 3)
plt.pie(expenses,
        labels=categories,
         autopct="%1.1f%%",colors=["skyblue","lightcoral","lightgreen"])
plt.title("Expenses Distribution")

plt.tight_layout()
plt.show()
```
